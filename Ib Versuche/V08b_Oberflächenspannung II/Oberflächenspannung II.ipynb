## **Messwerte**

```python
import numpy as np

#Aufgabe 1

u = 0.950e-3 #Umrechnungskonstante von V zu Pa in V/Pa

r = (3.85 / 2) * 10**(-3) #Radius der Kapilare in m
r_Fehler = 0.05e-3 / 2

deltap_wasser = 5.28 / u /74 #Druckdifferenz für Wasser deltaP = 5557.89 Pa

deltap_tensid = 3.92 / u /74 #Druckdifferenz für das Tensid deltaP = 4126.31 Pa

deltap_Fehler = (1/100) * deltap_wasser

deltap_Fehler_tensid = (1/100) * deltap_tensid

"""--------------------------------------------------------------------------------------------------------------"""

#Aufgabe 2

deltap_wasser_1 = 22.4 / u /74 #Druckdifferenz bei der unbekannten Kapilare in Wasser deltaP = 23578.94 Pa

deltap_tensid_1 = 12.4 / u  /74 #Druckdifferenz bei der unbekannten Kapilare im Tensid deltaP = 13052.63

deltap_Fehler_1 = (1/100) * deltap_wasser_1

deltap_Fehler_tensid_1 = (1/100) * deltap_tensid_1

```

## **Bestimmung der Oberflächenspannungen**

```python
import sympy
from IPython.display import display
from Skripte.Fehlerfortpflanzung import Gaußfehler

rp, dp = sympy.symbols("r, Delta_P")
expr1 = rp*dp/2
display(expr1)

gamma_wasser = 0.5*r*deltap_wasser

gamma_wasser_Fehler = Gaußfehler(expr1, np.array([rp, dp]), np.array([r, deltap_wasser]), np.array([r_Fehler, deltap_Fehler]))

print(f"\nOberflächenspannung von Wasser: gamma = {gamma_wasser} +/- {gamma_wasser_Fehler} N/m")


gamma_tensid = 0.5*r*deltap_tensid

gamma_tensid_Fehler = Gaußfehler(expr1, np.array([rp, dp]), np.array([r, deltap_tensid]), np.array([r_Fehler, deltap_Fehler_tensid]))

print(f"\nOberflächenspannung von Wasser mit Salz: gamma = {gamma_tensid} +/- {gamma_tensid_Fehler} N/m")
```

## **Bestimmung unbekannter Kapillarradius**

```python

gp, dp = sympy.symbols("gamma, Delta_P")
expr2 = 2*gp/dp
display(expr2)


r_wasser = 2*gamma_wasser/deltap_wasser_1

r_wasser_Fehler = Gaußfehler(expr2, np.array([gp, dp]), np.array([gamma_wasser, deltap_wasser_1]), np.array([gamma_wasser_Fehler, deltap_Fehler_1]))

print(f"\nBestimmter Radius in Wasser: r = {r_wasser*10**3} +/- {r_wasser_Fehler*10**3} mm")


r_tensid = 2*gamma_tensid/deltap_tensid_1

r_tensid_Fehler = Gaußfehler(expr2, np.array([gp, dp]), np.array([gamma_tensid, deltap_tensid_1]), np.array([gamma_tensid_Fehler, deltap_Fehler_tensid_1]))

print(f"\nBestimmter Radius in Salz: r = {r_tensid*10**3} +/- {r_tensid_Fehler*10**3} mm")
```

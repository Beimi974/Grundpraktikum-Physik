## **Messwerte**

```python
from math import gamma

import numpy as np

h_1 = np.array([18.3 , 17.8 , 20.5 , 21.7 , 18.3, 17.8, 20.6 , 19.4, 21.5, 18.4]) * 10**-2 #Erzeugter Höhe auf dem Manometer durch Überdruck

h_1_Fehler = 3e-3

"""-----------------------------------------------------------------------------------------------------------------------------------"""

h_2 = np.array([3.7 , 3.9 , 4.8 , 5.3 , 4.3, 4.6 , 5.3, 4.6, 4.9, 3.9]) *10**-2#Gemessene Höhe in m, wenn zu Umgebungstemperatur zurückgekehrt

h_2_Fehler = 3e-3

```

## **Bestimmung des Adiabatenexponenten**

```python
import sympy
from IPython.display import display
from Skripte.Fehlerfortpflanzung import Gaußfehler, VisualGaußfehler

h1, h2 = sympy.symbols("h_1, h_2")
expr1 =h1 / (h1 - h2)
display(expr1)

kappa = h_1 / (h_1 - h_2)

kappa_Fehler = np.ones(len(kappa), dtype= float)

for i in range(len(kappa)):
    kappa_Fehler[i] = Gaußfehler(expr1, np.array([h1, h2]), np.array([h_1[i], h_2[i]]), np.array([h_2_Fehler, h_1_Fehler]))
    print(f"\n #{i+1}: Adiabatenexponent: kappa = {kappa[i]} +/- {kappa_Fehler[i]}")

print(np.mean(kappa))
print(np.mean(kappa_Fehler))

display(VisualGaußfehler(expr1, np.array([h1, h2])))
```

## **Messwerte**

```python
import numpy as np

#Heizstrom I_Heiz = 1,5A

I_1_values = np.array([1, 1, 1, 1, 1, 1, 1, 1]) #Abgelesene Stromstärke über 4 Minuten alle 30 Sekunden
I_1 = np.mean(I_1_values)

I_1_Messfehler = 1

I_1_statfehler = np.std(I_1_values, ddof=1)/np.sqrt(8)
I_1_Fehler = np.sqrt(I_1_Messfehler**2 + I_1_statfehler**2)



U_1_values = np.array([1, 1, 1, 1, 1, 1, 1, 1]) #Abgelesene Spannung über 4 Minuten alle 30 Sekunden
U_1 = np.mean(U_1_values)

U_1_Messfehler = 1

U_1_statfehler = np.std(U_1_values, ddof=1)/np.sqrt(8)
U_1_Fehler = np.sqrt(U_1_Messfehler**2 + U_1_statfehler**2)


m_verdampft_1 = 1 #Masse des verdampften Wassers in kg
m_1_Fehler = 1

t_Mess_1 = 240 #Messzeit in Sekunden
t_Mess_1_Fehler = 0


"""---------------------------------------------"""

#Heizstrom I_Heiz = 1,8A

I_2_values = np.array([1, 1, 1, 1, 1, 1, 1, 1]) #Abgelesene Stromstärke über 4 Minuten alle 30 Sekunden
I_2 = np.mean(I_2_values)

I_2_Messfehler = 1

I_2_statfehler = np.std(I_2_values, ddof=1)/np.sqrt(8)
I_2_Fehler = np.sqrt(I_2_Messfehler**2 + I_2_statfehler**2)



U_2_values = np.array([1, 1, 1, 1, 1, 1, 1, 1]) #Abgelesene Spannung über 4 Minuten alle 30 Sekunden
U_2 = np.mean(U_2_values)

U_2_Messfehler = 1

U_2_statfehler = np.std(U_2_values, ddof=1)/np.sqrt(8)
U_2_Fehler = np.sqrt(U_2_Messfehler**2 + U_2_statfehler**2)


m_verdampft_2 = 1 #Masse des verdampften Wassers in kg
m_2_Fehler = 1

t_Mess_2 = 240 #Messzeit in Sekunden
t_Mess_2_Fehler = 0

"""------------------------------------------------------------------"""

#Heizstrom I_Heiz = 2A

I_3_values = np.array([1, 1, 1, 1, 1, 1, 1, 1]) #Abgelesene Stromstärke über 4 Minuten alle 30 Sekunden
I_3 = np.mean(I_3_values)

I_3_Messfehler = 1

I_3_statfehler = np.std(I_3_values, ddof=1)/np.sqrt(8)
I_3_Fehler = np.sqrt(I_3_Messfehler**3 + I_3_statfehler**3)



U_3_values = np.array([1, 1, 1, 1, 1, 1, 1, 1]) #Abgelesene Spannung über 4 Minuten alle 30 Sekunden
U_3 = np.mean(U_3_values)

U_3_Messfehler = 1

U_3_statfehler = np.std(U_3_values, ddof=1)/np.sqrt(8)
U_3_Fehler = np.sqrt(U_3_Messfehler**3 + U_3_statfehler**3)


m_verdampft_3 = 1 #Masse des verdampften Wassers in kg
m_3_Fehler = 1

t_Mess_3 = 240 #Messzeit in Sekunden
t_Mess_3_Fehler = 0

```

## **Berechnug der Verdampfungswärme**

```python
from Skripte.Fehlerfortpflanzung import Gaußfehler
import sympy
from IPython.display import display

Up, Ip, tp, mp = sympy.symbols("U, I, t, m")

expr1 = Up*Ip*tp / mp
display(expr1)

Variablen = np.array([Up, Ip, tp, mp])


L_1 = U_1*I_1*t_Mess_1 / m_verdampft_1

L_1_Fehler = Gaußfehler(expr1, Variablen, np.array([U_1, I_1, t_Mess_1, m_verdampft_1]), np.array([U_1_Fehler, I_1_Fehler, t_Mess_1_Fehler, m_1_Fehler]))

print(f"\n spezifische Verdampfungswärme von Wasser bei I = 1,5A: L = {L_1} +/- {L_1_Fehler} J/kg")



L_2 = U_2*I_2*t_Mess_2 / m_verdampft_2

L_2_Fehler = Gaußfehler(expr1, Variablen, np.array([U_2, I_2, t_Mess_2, m_verdampft_2]), np.array([U_2_Fehler, I_2_Fehler, t_Mess_2_Fehler, m_2_Fehler]))

print(f"\n spezifische Verdampfungswärme von Wasser bei I = 1,8A: L = {L_2} +/- {L_2_Fehler} J/kg")


L_3 = U_3*I_3*t_Mess_3 / m_verdampft_3

L_3_Fehler = Gaußfehler(expr1, Variablen, np.array([U_3, I_3, t_Mess_3, m_verdampft_3]), np.array([U_3_Fehler, I_3_Fehler, t_Mess_3_Fehler, m_3_Fehler]))

print(f"\n spezifische Verdampfungswärme von Wasser bei I = 2A: L = {L_3} +/- {L_3_Fehler} J/kg")




```

```python
#Automatische Berechnung Fehlerfortpflanzung nach Gauß

def Gausfehler(expr, Variablen, Mittelwerte, Fehler):
    """
    :param Fehler:
    :param Mittelwerte:
    :param expr:
    :param Variablen:
    :return:
    """

    import sympy
    import numpy as np
    from IPython.display import display

    count = 0
    for pos in range(len(Variablen)): #Loop durch alle fehlerbehafteten Variablen

        deriv = expr.diff(Variablen[pos]) #Ableitung nach der aktuellen Variablen

        deriv_func = sympy.lambdify(Variablen, deriv, "numpy") #Ableitung in aufrufbare Funktion umwandeln

        count += (Fehler[pos] * deriv_func(*Mittelwerte)) ** 2 #Nach der Formel (Fehler * Ableitung)**2 addieren

    error = np.sqrt(count) #Nach Formel Wurzel aus der bestimmten Summe ziehen

    #Ausgabe
    print("Term für den Fehler bestimmt wird:\n")
    display(expr)
    print("")
    print(f"Fehler: {error: 0.10E}")
```

```python
import sympy
import numpy as np
from IPython.display import display

#Automatische Berechnung Fehlerfortpflanzung nach Gauß

#Konstanten
g = 9.81
pi = np.pi

#Fehlerbehaftete Variablen
L, Rprime, a, R = sympy.symbols("L,R',a,R")
expr = sympy.Function("expr")

#Term zur Berechnung
expr = (2 * g * L * Rprime) / (pi * a * (R ** 4))

#Variablen, Mittelwerte und zugehörige Fehler
# Variable, Mittelwert, Fehler
Werte = np.array([[L, 95.4e-2, 2e-3],  #Lönge in m
                  [Rprime, 8.2e-2, 1e-3],  # Scheibendurchmesser in m
                  [a, 8.7522, 0.12196],  #Bestimmte Steigung
                  [R, 2.43e-3 / 2, 0.01e-3]  #Radius des Drahts
                  ])

count = 0
for pos in range(len(Werte[:, 0])):  #Loop durch alle fehlerbehafteten Variablen

    deriv = expr.diff(Werte[pos, 0])  #Ableitung nach der aktuellen Variablen

    deriv_func = sympy.lambdify(Werte[:, 0], deriv, "numpy")  #Ableitung in aufrufbare Funktion umwandeln

    count += (Werte[pos, 2] * deriv_func(*Werte[:, 1])) ** 2  #Nach der Formel (Fehler * Ableitung)**2 addieren

Error = np.sqrt(count)  #Nach Formel Wurzel aus der bestimmten Summe ziehen

#Ausgabe
print("Term für den Fehler bestimmt wird:\n")
display(expr)
print("")
print(f"Fehler: {Error: 0.10E} Pa")
```

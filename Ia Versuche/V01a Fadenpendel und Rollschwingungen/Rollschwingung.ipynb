## **Messwerte**

```python
import numpy as np
import sympy as sp
from Skripte.Fehlerfortpflanzung import Gaußfehler

g = 9.81

m_kugel = 32.16e-3 #Masse der Kugel in kg
m_zylinder = 11.88e-3 #Masse des Zylinders in kg
m_Hzylinder = 6.65e-3 #Masse des Hohlzylinders in kg
m_Fehler = 0.01e-3

r_kugel = (2.0/ 2) * 10**(-2) #Radius der Kugel in m
r_zylinder = (2.0 / 2) *10**(-2) #Radius des Zylinders in m

ra_Hzylinder = 3.0 / 2 * 10**(-2) #Außenradius des Hohlzylinders in m
ri_Hzylinder = 2.6 / 2 * 10**(-2) #Innenradius des Hohlzylinders in m

r_Fehler = 0.05e-3 #Fehler der Messungen mit der Schiebelehre

R = 0.5156 #Krümmungsradius des Uhrenglases in m

d, h = sp.symbols("d, h")
expr_0 = h/2 + d**2/(8*h)
R_Fehler = Gaußfehler(expr_0, np.array([d, h]), np.array([29.8e-2, 2.2e-2]), np.array([3e-3, 0.05e-3]))
#print(R_Fehler)

t_kugel = np.array([8.66, 8.40, 8.21, 8.61, 8.53]) #Zeiten für jeweils 5 Schwingungen der Kugel im Uhrenglas

t_zylinder = np.array([8.80, 8.76, 8.90, 8.73, 8.61]) #Zeiten für jeweils 5 Schwingungen des Zylinders im Uhrenglas

t_Hzylinder = np.array([9.69, 9.84, 9.56, 9.54, 9.58]) #Zeiten für jeweils 5 Schwingungen des Hohlzylinders im Uhrenglas

t_Fehler = 0.1

T_k = np.mean(t_kugel/5)
T_z = np.mean(t_zylinder/5)
T_Hz = np.mean(t_Hzylinder/5)

T_Fehler = np.sqrt( (t_Fehler/5)**2 + ((np.std(t_zylinder, ddof= 1) / np.sqrt(5))**2) )
```

## **Trägheitsmoment Kugel**

```python
import sympy as sp
from Skripte.Fehlerfortpflanzung import Gaußfehler

I_Kugel = 2/5 * m_kugel * r_kugel**2

mk, rk = sp.symbols("m_k, r_k")
expr_1 = sp.Rational(2,5) * mk * rk**2
display(expr_1)

Variablen_1 = np.array([mk, rk])
Mittelwerte_1 = np.array([m_kugel, r_kugel])
Fehler_1 = np.array([m_Fehler, r_Fehler])

err_1 = Gaußfehler(expr_1, Variablen_1, Mittelwerte_1, Fehler_1)

print(f"Trägheitsmoment Kugel berechnet: I_k = {I_Kugel} +/- {err_1} kg*m**2")
```

## **Trägheitsmoment Zylinder**

```python
import sympy as sp
from Skripte.Fehlerfortpflanzung import Gaußfehler

I_Zylinder = 1/2 * m_zylinder * r_zylinder**2

mz, rz = sp.symbols("m_z, r_z")
expr_2 = sp.Rational(1,2) * mz * rz**2
display(expr_2)

Variablen_2 = np.array([mz, rz])
Mittelwerte_2 = np.array([m_zylinder, r_zylinder])
Fehler_2 = np.array([m_Fehler, r_Fehler])

err_2 = Gaußfehler(expr_2, Variablen_2, Mittelwerte_2, Fehler_2)

print(f"Trägheitsmoment Zylinder berechnet: I_Z = {I_Zylinder} +/- {err_2} kg*m**2")
```

## **Trägheitsmoment Hohlzylinder**

```python
import sympy as sp
from Skripte.Fehlerfortpflanzung import Gaußfehler

I_Hzylinder = 1/2 * m_Hzylinder * (ra_Hzylinder**2 + ri_Hzylinder**2)

mhz, rih, rah = sp.symbols("m_hz, r_i, r_a")
expr_3 = sp.Rational(1,2) * mhz * (rih**2 + rah**2)
display(expr_3)

Variablen_3 = np.array([mhz, rih, rah])
Mittelwerte_3 = np.array([m_Hzylinder, ri_Hzylinder, ra_Hzylinder])
Fehler_3 = np.array([m_Fehler, r_Fehler, r_Fehler])

err_3 = Gaußfehler(expr_3, Variablen_3, Mittelwerte_3, Fehler_3)

print(f"Trägheitsmoment Hohlzylinder berechnet: I_HZ = {I_Hzylinder} +/- {err_3} kg*m**2")
```

## **Radius des Uhrenglases**

```python
import sympy as sp
from IPython.display import display

R_b = r_kugel + (5 * g * T_k**2) / (28*np.pi**2)

rp, Tp = sp.symbols("r_k, T_R")
expr_4 = rp + sp.Rational(5,28) * (g*Tp**2) / sp.pi**2
display(expr_4)

Variablen_4 = np.array([rp, Tp])
Mittelwerte_4 = np.array([r_kugel, T_k])
Fehler_4 = np.array([r_Fehler, T_Fehler])

err_4 = Gaußfehler(expr_4, Variablen_4, Mittelwerte_4, Fehler_4)

print(f"Radius Uhrenglas berechnet: R = {R_b} +/- {err_4} m   (gemessen: {R}m)")
```

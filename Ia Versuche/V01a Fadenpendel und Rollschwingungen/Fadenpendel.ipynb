## **Messungen**

```python
import numpy as np

m = 17.50e-3 #Masse des Pendelkörpers in kg
m_Fehler = 0.01e-3

R = 1.59e-2 /2 #Radius der Kugel des Fadenpendels in m
R_Fehler = 0.05e-3 /2

l = 81.5e-2 + R #Länge des Fadenpendels in m
l_Fehler = 3e-3

phi_0 = 7.147e-2 #Anfangswinkel des Pendels in Radians
phi_0_Fehler = 0

rho_Luft = 1.225 #Dichte von Luft in kg/m**3

t = np.array([91044, 90727, 90742, 90747, 90780, 91039, 90761, 91101, 91086, 90738, 90993, 90995]) / 1000 #gemessene Zeiten für 50 Schwinungen des Pendels in s
t_Fehler = np.std(t, ddof = 1) / np.sqrt(12)

T = np.mean(t/50)
T_Fehler = t_Fehler/50 #Fehler für T nur 1/50 des von t (Gauß)
```

## **Berechnung ohne Korrektur**

```python
from Skripte.Fehlerfortpflanzung import Gaußfehler
from IPython.display import display
import sympy as sp

g = (l*4*(np.pi**2)) / (T**2)

lp, Tp =sp.symbols("l,T")
expr = (lp*4*(sp.pi**2)) / (Tp**2)
display(expr)

Variablen = np.array([lp,Tp])
Fehler = np.array([l_Fehler, T_Fehler])
Mittelwerte = np.array([l,T])

err = Gaußfehler(expr, Variablen, Mittelwerte, Fehler)

print(f"g = {g} +/- {err}")
```

## **Korrektur Auslenkung (Winkelnäherung)**

```python
from Skripte.Fehlerfortpflanzung import Gaußfehler
from IPython.display import display
import sympy as sp

Tprime = T*(1+ 1/16 * phi_0**2)

g_1 = (l*4*(np.pi**2)) / (Tprime**2)

lp, Tp, phi = sp.symbols("l,T,phi_0")
expr_1 = (lp*4*(sp.pi**2)) / ( ( Tp*(1 + sp.Rational(1,16)* phi**2) )**2 )
display(expr_1)

Variablen_1 = np.array([lp,Tp, phi])
Fehler_1 = np.array([l_Fehler, T_Fehler, phi_0_Fehler])
Mittelwerte_1 = np.array([l, T, phi_0])

err_1 = Gaußfehler(expr_1, Variablen_1, Mittelwerte_1, Fehler_1)

print(f"g_1 = {g_1} +/- {err_1}")
```

## **Korrektur Trägheitsmoment (Näherung physikalisches Pendel)**

```python
from Skripte.Fehlerfortpflanzung import Gaußfehler
from IPython.display import display
import sympy as sp

lprime = l*(1 + 2/5 * (R**2)/(l**2) )

g_2 = (lprime*4*(np.pi**2)) / (T**2)

lp, Rp, Tp = sp.symbols("l,R,T")
epxr_2 = ( (lp*(1 + sp.Rational(2,5) * (Rp**2)/(lp**2))) * 4 * (sp.pi**2)) / (Tp**2)
display(epxr_2)

Variablen_2 = np.array([lp, Rp, Tp])
Fehler_2 = np.array([l_Fehler, R_Fehler, T_Fehler])
Mittelwerte_2 = np.array([l, R, T])

err_2 = Gaußfehler(epxr_2, Variablen_2, Mittelwerte_2, Fehler_2)

print(f"g_2 = {g_2} +/- {err_2}")
```

## **Korrektur Auftrieb in Luft**

```python
from Skripte.Fehlerfortpflanzung import Gaußfehler
from IPython.display import display
import sympy as sp

m_Luft = rho_Luft * 4/3 * np.pi * R**3

g_3 = (l*4*(np.pi**2)) / (T * 1/(np.sqrt(1 - m_Luft/m) ))**2

lp, Rp, Tp, mp, ml = sp.symbols("l, R, T, m, m_L")

expr_3 = (lp * 4 * sp.pi**2) / (Tp * (1/ sp.sqrt(1 - (ml /  mp)) ) )**2
display(expr_3)

Variablen_3 = np.array([lp, Rp, Tp, mp, ml])
Fehler_3 = np.array([l_Fehler, R_Fehler, T_Fehler, m_Fehler, 0])
Mittelwerte_3 = np.array([l, R, T, m, m_Luft])

err_3 = Gaußfehler(expr_3, Variablen_3, Mittelwerte_3, Fehler_3)

print(f"g_3 = {g_3} +/- {err_3}")

g_4 = (lprime*4*(np.pi**2)) / (Tprime * 1/(np.sqrt(1 - m_Luft/m) ))**2
print(g_4)
```
